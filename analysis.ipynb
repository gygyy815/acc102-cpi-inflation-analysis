## Data Source
- Source: U.S. Bureau of Labor Statistics - Consumer Price Index for All Urban Consumers (CPI-U)
- Access Date: [2026/4/22]
- Data File: cpiai.csv
- Key Fields: Date (monthly), Index (CPI Index value), Inflation (monthly inflation rate)

Step 1: Import the required libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12,6)
```

Step 2: Load the CSV dataset

```python
import os
data_path = os.path.join("data", "cpiai.csv")
df = pd.read_csv(data_path)
df.head()
```

Step 3: Convert the 'Date' column to datetime

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Check for any invalid dates
print("Number of invalid dates:", df['Date'].isnull().sum())
```

Step 4: Extract year and filter 2010-2023

```python
# Create a 'Year' column from the Date field
df['Year'] = df['Date'].dt.year

# Filter data for years 2010 to 2023 inclusive
df_filtered = df[(df['Year'] >= 2010) & (df['Year'] <= 2023)]

# Remove any rows with invalid dates
df_filtered = df_filtered.dropna(subset=['Date'])

# Display first few rows to verify
print(f"Filtered data shape: {df_filtered.shape}")
df_filtered.head()
```

Step 5: Plot the monthly CPI inflation trend

```python
plt.figure()
sns.lineplot(data=df_filtered, x='Date', y='Inflation')
plt.title("US CPI Inflation Trend (2010-2023)")
plt.ylabel("Inflation (%)")
plt.xlabel("Date")
plt.show()
```

Step 6: Calculate the annual average inflation and plot a bar chart

```python
annual = df_filtered.groupby('Year')['Inflation'].mean().reset_index()

plt.figure()
sns.barplot(data=annual, x='Year', y='Inflation', palette="Blues_d")
plt.title("Average Annual US CPI Inflation (2010-2023)")
plt.ylabel("Inflation (%)")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.show()

sns.barplot(data=annual, x='Year', y='Inflation', palette="Blues_d", hue=None)
```

Step 7: Print key summary statistics

```python
max_inflation = df_filtered['Inflation'].max()
min_inflation = df_filtered['Inflation'].min()
year_max_avg = annual.loc[annual['Inflation'].idxmax(), 'Year']

print(f"Maximum monthly inflation: {max_inflation:.2f}%")
print(f"Minimum monthly inflation: {min_inflation:.2f}%")
print(f"Year with highest annual average inflation: {year_max_avg}")
```
